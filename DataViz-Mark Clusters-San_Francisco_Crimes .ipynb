# San Francisco Crimes Map Markers

Request of an interactive map with markers and labels of crimes registered in San Francisco in 2016. 

Then, create an interactive map with mark clusters.

Dataset: https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv'


- Use 100 samples


```python
import numpy as np
import pandas as pd
#!pip install geopy
from geopy.geocoders import Nominatim
#!conda install -c conda-forge folium=0.5.0 --yes
import folium
from folium import plugins
```

```python
df_incidents = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
print(df_incidents.shape,'\n')
print('Null values:',df_incidents['Category'].isnull().sum(),'\n')
df_incidents.head()
```

```python
# First 100 samples
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
# Get latitude, longitude
address = 'San Francisco'
geolocator = Nominatim(user_agent="sanfran_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

```python
# Plain map
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)
sanfran_map
```

```python
# Adding data and markers
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.features.CircleMarker(
        [lat, lng],
        radius=5, 
        color='yellow',
        fill=True,
        popup=label, #df_incidents['Category']
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)

sanfran_map
```

```python
# Clean copy of San Francisco Map
sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

# Mark cluster object for data
incidents = plugins.MarkerCluster().add_to(sanfran_map)

# Adding data to mark clusters
for lat, lng, label, in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
sanfran_map
```
